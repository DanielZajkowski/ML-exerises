###  Process data

```python
import os
import pandas as pd
import xport
import csv
filenames = ['h181', 'h192']

for file in filenames:
    CSV_FILE_NAME = file+".csv"
    SSP_FILE_NAME = file+".ssp"

    base_path = "/opt/conda/envs/Python-3.7-main/lib/python3.7/site-packages/aif360/data/raw/meps/"
    csv_file_path = os.path.join(base_path, CSV_FILE_NAME)
    ssp_file_path = os.path.join(base_path, SSP_FILE_NAME)

    if not os.path.isfile(csv_file_path):
        if os.path.isfile(ssp_file_path):
            with open(ssp_file_path, 'rb') as in_xpt:
                print("opened ", ssp_file_path)
                with open(csv_file_path, 'w', newline='') as out_csv:
                    print("opened ", csv_file_path)
                    reader = xport.Reader(in_xpt)
                    print(reader)
                    writer = csv.writer(out_csv, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                    writer.writerow(reader.fields)
                    for row in reader:
                        writer.writerow(row)
```
